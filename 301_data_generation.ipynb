# Create data for Gelateria task 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123456)


def make_data_point(mu, sigma, n):
    data = np.random.normal(mu, sigma, n)
    data = np.where(data < 1.0, 1.0, data)
    data = np.where(data > 5.0, 5.0, data)
    data = np.round(data)
    return data


def make_dataset(props, filename):
    NUM = 30
    data = {}
    for key, vals in props.items():
        data[key] = make_data_point(*vals, NUM)

    pure_data = np.array(list(data.values())).T
    header = ", ".join([str(element) for element in data.keys()])
    np.savetxt(filename, pure_data, header=header)
```
